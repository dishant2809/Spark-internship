```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Ship Mode'].unique()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(df['State'].value_counts().index,df['State'].value_counts(),palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(df['Category'].value_counts().index,df['Category'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(df['Sub-Category'].value_counts(),labels=df['Sub-Category'].value_counts().index)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(df['Category'],hue=df['Segment'])
plt.show()
```

```python
df.hist(figsize=(10,10),bins=50)
plt.show()
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['orange','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```
