```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns',150)
pd.set_option('display.max_rows',150)
```

```python
data = pd.read_csv('glob_terr_index.csv',encoding='latin1')
data.head()
```

```python
df = data[['iyear','imonth','iday','country_txt','region_txt','success','suicide','targtype1_txt','gname','weaptype1_txt']]
```

```python
df.head()
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
plt.figure(figsize=(25,12))
sns.countplot(x=df['iyear'],hue='success',palette='inferno',data=df)
plt.title('Attack by year with success or fail',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(y=df['weaptype1_txt'].value_counts().values[:5], x=df['weaptype1_txt'].value_counts().index[:5])
plt.title('Top 5 Weapon Type used',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.barplot(y=df['country_txt'].value_counts().values[:20], x=df['country_txt'].value_counts().index[:20])
plt.xticks(rotation=90)
plt.title('20 most affected country',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.barplot(y=df['targtype1_txt'].value_counts().values[:15], x=df['targtype1_txt'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.title('15 most target places',fontsize=25)
plt.show()
```

```python
plt.figure(figsize=(20,15))
sns.barplot(y=df['gname'].value_counts().values[:15], x=df['gname'].value_counts().index[:15])
plt.xticks(rotation=90)
plt.title('15 most active terror orginisation',fontsize=25)
plt.show()
```

```python
plt.rcParams['figure.figsize']=15,5
a=pd.crosstab(df['iyear'],df['region_txt'])
a.plot()
plt.title('Region of attack',fontsize=25)
```
